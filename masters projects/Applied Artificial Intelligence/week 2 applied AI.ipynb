```python
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error,accuracy_score, classification_report, confusion_matrix
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
GNB = GaussianNB()
RFC = RandomForestClassifier()
DT = DecisionTreeClassifier()
```

```python
#getting the x and y axis
x =  cefas1[['Alexandrium spp. cells L-1 (PSP)','Pseudo-nitzschia spp. cells L-1 (ASP) ', 
             'Dinophysiaceae cells L-1(DSP)','Prorocentrum lima cells L-1 (DSP)']]
y = cefas1[['thresholds_reached']]
```

```python
#using minmax scaler
scaler= MinMaxScaler()
x_scaled = scaler.fit_transform(x)
y_scaled = scaler.fit_transform(y)
```

```python
print(x_scaled)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x_scaled, y_scaled, test_size = 0.2, random_state = 42)
```

```python
print(len(xtrain), len(xtest), len(ytest), len(ytrain))
```

### GAUSSIAN NAIVE BAYES

```python
GNB.fit(xtrain, ytrain)
```

```python
GNB_pred = GNB.predict(xtest)
```

```python
GNB_result = classification_report(ytest, GNB_pred)
print(GNB_result)
```

```python
test = accuracy_score(GNB_pred, ytest)
print(test)
```

```python
GNB_confmatr = confusion_matrix(ytest, GNB_pred)
print(GNB_confmatr)
```
