```python
# function for linear regression
def Linear(xtest, xtrain, ytest, ytrain):
    from sklearn.linear_model import LinearRegression # importing the linear regression library
    from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score # importing the evaluation metrics
    
    Regression = LinearRegression() # instance of the model
    Regression.fit(xtrain, ytrain) # training the model with the train data
    pred_test = Regression.predict(xtest)
    
    # getting the training error
    pred_train = Regression.predict(xtrain)
    e = mean_squared_error(ytrain,pred_train)
    
    
    a= mean_absolute_error(ytest,pred_test)
    b = mean_squared_error(ytest,pred_test) # getting the test error
    c= np.sqrt(b)
    d= r2_score(ytest,pred_test)
    o = abs(e-b) # overfitting = (train error - test error)
    return(a, b, c, d, o)

#function to check for errors
def print_errors(data, ma, ms, rmse, cd, err):
    print(f'###############  {data} ###############')
    print(f"The result of the mean absolute error is: {ma:.2f} ")
    print(f"The result of the mean squared error is : {ms:.2f} using NA_Sales column")
    print(f"The result of the root mean squared error is : {rmse:.2f} using NA_Sales column")
    print(f"The result of the coefficient of determination is: {cd:.2f} using NA_Sales column")

    if err < 0.5:
        print(f'The model is not overfitting with a value of {err}' )
    else: 
        print(f'The model is overfitting with a value of {err}')
```

```python
input_data = 'NA_Sales' # NA SALES
mean_abs, mean_squared, root_mean_sq_err, coef_det, check_fitting = Linear(test_xval_NASales, trained_xval_NASales, test_yval, trained_yval )
print_errors(input_data, mean_abs, mean_squared, root_mean_sq_err, coef_det, check_fitting)
```

```python
#####################TRAIN TEST SPLIT##############################
def reform_data(X_data, y_data):
    def bal_data(X_data, y_data):
       # resample data using smote
        from imblearn.over_sampling import SMOTE
        try:
            sm = SMOTE(random_state=42, k_neighbors = 2)
            bal_x, bal_y = sm.fit_resample(X_data, y_data)
        except NameError as err: 
            print('Some values are not defined', err)
        except ValueError as err: 
            print('All values should be integers', err)
        return bal_x, bal_y
    
    # split the data to 80/20
    def split_data(bal_x, bal_y):
        from sklearn.model_selection import train_test_split
        trained_xval, test_xval, trained_yval, test_yval = train_test_split(bal_x, bal_y, test_size=0.2, random_state=42)
        return trained_xval, test_xval, trained_yval, test_yval
    
    # using min max scaler to scale the data
    def scale_data(trained_xval, test_xval):
        from sklearn.preprocessing import MinMaxScaler
        scaler = MinMaxScaler()
        trained_xval_scaled = scaler.fit_transform(trained_xval)
        test_xval_scaled = scaler.transform(test_xval)
        return trained_xval_scaled,test_xval_scaled
   
    # label encoding for decision tree
    def label_encode(data):
        from sklearn.preprocessing import LabelEncoder     
        label_encoder = LabelEncoder()
        data_encoded = label_encoder.fit_transform(data)
        return data_encoded
    
    y_encoded = label_encode(y_data)
    bal_x, bal_y = bal_data(X_data, y_data)
    trained_xval, test_xval, trained_yval, test_yval = split_data(bal_x, bal_y)
    trained_xval_scaled,test_xval_scaled = scale_data(trained_xval, test_xval)    
    return trained_xval_scaled, test_xval_scaled, trained_yval, test_yval, y_encoded 
    
```

```python
#################### USING CROSS VALIDATION###################################################
def reform_data_cv(X_data, y_data):
    def bal_data(X_data, y_data):
        from imblearn.over_sampling import SMOTE
        sm = SMOTE(random_state=42, k_neighbors = 2)
        bal_x, bal_y = sm.fit_resample(X_data, y_data) # The object is applied
        return bal_x, bal_y
    
    def scale_data(bal_x):
        from sklearn.preprocessing import MinMaxScaler
        scaler = MinMaxScaler()
        X_scaled = scaler.fit_transform(bal_x)
        return X_scaled
    
    bal_x, bal_y = bal_data(X_data, y_data)
    X_scaled = scale_data(bal_x)
    
    return X_scaled, bal_y
    
```

```python
def classify_cv(X_scaled, bal_y): # x and y data using cross validation for comparison with train test split
    from sklearn.metrics import classification_report, confusion_matrix
    from sklearn.model_selection import cross_val_score
    from sklearn.model_selection import StratifiedKFold 
    import warnings
    warnings.simplefilter(action='ignore', category=FutureWarning)
    cross_val = StratifiedKFold(n_splits=7, shuffle=True, random_state=42)
    
    def logReg(X_scaled, bal_y):
        from sklearn.linear_model import LogisticRegression
        model = LogisticRegression()
        scores = cross_val_score(model, X_scaled, bal_y, cv=cross_val)
        model.fit(X_scaled, bal_y)
        pred_log = model.predict(X_scaled)
        log_result = classification_report(bal_y, pred_log)
        return scores, log_result
    
    def svc(X_scaled, bal_y):
        from sklearn.svm import SVC
        model = SVC()
        scores = cross_val_score(model, X_scaled, bal_y, cv=cross_val)
        model.fit(X_scaled, bal_y)
        pred_svc = model.predict(X_scaled)
        svc_result = classification_report(bal_y, pred_svc)
        return scores, svc_result
    
    def KNN(X_scaled, bal_y):
        from sklearn.neighbors import KNeighborsClassifier
        model = KNeighborsClassifier()
        scores = cross_val_score(model, X_scaled, bal_y, cv=cross_val)
        model.fit(X_scaled,bal_y)
        pred_knn = model.predict(X_scaled)
        knn_result = classification_report(bal_y, pred_knn)
        return scores, knn_result
    
    def Gaussian(X_scaled, bal_y):
        from sklearn.naive_bayes import GaussianNB
        model = GaussianNB()
        scores = cross_val_score(model, X_scaled, bal_y, cv=cross_val)
        model.fit(X_scaled, bal_y)
        pred_gaussian = model.predict(X_scaled)
        gaussian_result = classification_report(bal_y, pred_gaussian)
        return scores, gaussian_result
    
    a1, a2 =  logReg(X_scaled, bal_y)  
    print('################## LOGISTIC REGRESSION #######################')
    print('Cross-validation scores: ', a1)
    print('Mean score: ', a1.mean())
    print('Standard deviation: ', a1.std())
    print(a2)
    
    b1, b2 = svc(X_scaled, bal_y)
    print('################## SUPPORT VECTOR MACHINE #######################')
    print('Cross-validation scores: ', b1)
    print('Mean score: ', b1.mean())
    print('Standard deviation: ', b1.std())
   
    print('Support Vector Classification Report:')
    print(b2)

    c1, c2 = KNN(X_scaled, bal_y)
    print('################## K NEAREST NEIGHBOR #######################')
    print('Cross-validation scores: ', c1)
    print('Mean score: ', c1.mean())
    print('Standard deviation: ', c1.std())
    
    print('Knearest Classification Report:')
    print(c2)

    d1, d2 = Gaussian(X_scaled, bal_y)
    print('################## GAUSSIAN #######################')
    print('Cross-validation scores: ', d1)
    print('Mean score: ', d1.mean())
    print('Standard deviation: ', d1.std())
    print('Gaussian Classification Report:')
    print(d2)
```
