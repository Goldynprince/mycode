# Exercise 3


```python
coordinates= []
for i in range(100):
    x = (round(random.uniform(1.0,100.0),1))
    y = (round(random.uniform(1.0,100.0),1)) 
    coordinates.append((x,y))

print(coordinates)
```

# Exercise 4

I would expect the line of best fit to be in the mid point of the chart. 

```python
import numpy as np
import matplotlib.pyplot as plt

X = []
Y = []
for coordinate in coordinates:
    a, b = coordinate
    X.append(a)
    Y.append(b)
    
x = np.array(X)
y =np.array(Y)
```

```python
plt.plot(y, x, 'ko', markerfacecolor = "y", markersize = 8)
plt.xlim(-10,110)
plt.ylim(-10,110)
plt.xlabel("x coordinates")
plt.ylabel("y coordinates")

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mymodel = LinearRegression()
mymodel.fit(x.reshape(-1, 1), y)
```

```python
print("Intercept, c = ", mymodel.intercept_)
print("Gradient, m = ", mymodel.coef_)
```

```python
coordinate_pred = mymodel.intercept_ + mymodel.coef_*x
print("Predictions:", coordinate_pred)
```

```python
plt.plot(y, x, 'ko', markerfacecolor = "y", markersize = 8)
plt.plot(x,coordinate_pred, 'r-', linewidth = 1.8)

plt.xlim(-10,110)
plt.ylim(-10,110)
plt.xlabel("x coordinates")
plt.ylabel("y coordinates")

plt.show()
```

```python
# First import the following modules: 
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.optimize import leastsq 
# Now we have to define some function that describes a parabola: 

def func(params, x): 
    a, b, c = params 
    return a * x * x + b * x + c 


# Define an error function: 
def error(params, x, y): 
    return func(params, x) - y 


# And a function to find the solution parameters: 
def solvePara(): 
    p0 = [10, 10, 10] 
    Para = leastsq(error, p0, args=(X, Y)) 
    return Para 

# Finally, the solution can be done as follows: 
def solution(): 
    Para = solvePara() 
    a, b, c = Para[0] 
    print("a=",a," b=",b," c=",c) 
    print("The equation of the curve is:") 
    print("y="+str(round(a,2))+"x*x+"+str(round(b,2))+"x+"+str(c))
solution()
```
