```python
# add up values till max and sum up rest
def get_sellable(needed_hydrogen, data):
    supplied_hydrogen = 0
    excess = 0
    
    for index, hydrogen in enumerate(data): 
        supplied_hydrogen = supplied_hydrogen + hydrogen
        if supplied_hydrogen > needed_hydrogen:
            excess = data[index +1:].sum()# the index when the supplied is met. we add the rest for the day
            return (index, supplied_hydrogen, excess)
        
        
def get_data(day):#get a temp data with the hourly information
    start_index = day*24
    end_index = start_index+24
    temp_data = hourly['1hour_hydrogen_forecast'][start_index:end_index]# data for the day being queried
    return temp_data
```

```python
tolerance = 1000
hydrogen_demand = daily['Hydrogen_Demand_Forecast_KG']
hydrogen_forecast = daily['daily_hydrogen_forecast']
dat = []# holds dtaa for if we need hydrogen or we have excess and how much is needed or excess. 0 is excess, 1 is needed
for i in range(365):
    hydrogen_demand = daily['Hydrogen_Demand_Forecast_KG'][i]
    hydrogen_forecast = daily['daily_hydrogen_forecast'][i]
    if (hydrogen_demand +tolerance) < hydrogen_forecast: 
        dat.append((0, i+1, get_sellable(hydrogen_demand, get_data(i))))# profit or loss, the day,the index we met demand,  the hydrogen supplied and the excess
    else: 
        shortage_hydrogen = (hydrogen_demand +tolerance) - hydrogen_forecast
        dat.append((1, i+1,shortage_hydrogen ))# profit or loss, day, needed hydrogen
        
        
        
```

```python
def get_price_time(start, end):# get the lowest price of hydrogen to buy at and the time to start buying. 
    temp_data = hourly[['DateHour','Electricity_Prices_Day_Ahead_Dollar']][start:end] 
    temp_data2 = hourly['Electricity_Prices_Day_Ahead_Dollar'][start:end] 
    lowest_electricity_cost = temp_data.loc[temp_data2.idxmin()].Electricity_Prices_Day_Ahead_Dollar
    time_of_low_elec = temp_data.loc[temp_data2.idxmin()].DateHour

    return (time_of_low_elec, lowest_electricity_cost)
            
def get_profit_loss(data):
    profit =[]
    loss =[]
    for values in data:
        start_index = values[1]*24
        end_index = start_index +23
        if values[0]== 0: # we have excess
            
            #profit is the revenue for every hour we sell wind energy
            temp_data = hourly['1hour_MWH_Sellinprice_forecast'][start_index:end_index]
            profit.append(temp_data[values[2][0] +1:].sum())# add up the price of selling electricity to the market at hourly price 
        elif values[0] == 1: # we need hydrogen 
         #loss if the cost of electricity for every hour we need hydrogen
            time, price = get_price_time(start_index, end_index)
            loss.append((time, (values[2] * price)+(0.1*price)))# time we start buying hydrogen and the cost we have to spend to meet day demand plus overhead cost

    return profit, loss      
```
