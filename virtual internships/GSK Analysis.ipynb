```python
#################### USING CROSS VALIDATION###################################################
def reform_data_cv(X_data, y_data):
    def bal_data(X_data, y_data):
        from imblearn.over_sampling import SMOTE
        sm = SMOTE(random_state=42, k_neighbors = 2)
        bal_x, bal_y = sm.fit_resample(X_data, y_data) # The object is applied
        return bal_x, bal_y
    
    def scale_data(bal_x):
        from sklearn.preprocessing import MinMaxScaler
        scaler = MinMaxScaler()
        X_scaled = scaler.fit_transform(bal_x)
        return X_scaled
    
    bal_x, bal_y = bal_data(X_data, y_data)
    X_scaled = scale_data(bal_x)
    
    return X_scaled, bal_y
    
```

```python
def classify_cv(X_scaled, bal_y): # x and y data using cross validation for comparison with train test split
    from sklearn.metrics import classification_report, confusion_matrix
    from sklearn.model_selection import cross_val_score
    from sklearn.model_selection import StratifiedKFold 
    import warnings
    warnings.simplefilter(action='ignore', category=FutureWarning)
    cross_val = StratifiedKFold(n_splits=7, shuffle=True, random_state=42)
    
    def logReg(X_scaled, bal_y):
        from sklearn.linear_model import LogisticRegression
        model = LogisticRegression()
        scores = cross_val_score(model, X_scaled, bal_y, cv=cross_val)
        model.fit(X_scaled, bal_y)
        pred_log = model.predict(X_scaled)
        log_result = classification_report(bal_y, pred_log)
        log_confmatr = confusion_matrix(bal_y, pred_log)
        return scores, log_result, log_confmatr
    
    def svc(X_scaled, bal_y):
        from sklearn.svm import SVC
        model = SVC()
        scores = cross_val_score(model, X_scaled, bal_y, cv=cross_val)
        model.fit(X_scaled, bal_y)
        pred_svc = model.predict(X_scaled)
        svc_result = classification_report(bal_y, pred_svc)
        svc_confmatr = confusion_matrix(bal_y, pred_svc)
        return scores, svc_result, svc_confmatr
    
    def KNN(X_scaled, bal_y):
        from sklearn.neighbors import KNeighborsClassifier
        model = KNeighborsClassifier()
        scores = cross_val_score(model, X_scaled, bal_y, cv=cross_val)
        model.fit(X_scaled,bal_y)
        pred_knn = model.predict(X_scaled)
        knn_result = classification_report(bal_y, pred_knn)
        knn_confmatr = confusion_matrix(bal_y, pred_knn)
        return scores, knn_result, knn_confmatr
    
    def Gaussian(X_scaled, bal_y):
        from sklearn.naive_bayes import GaussianNB
        model = GaussianNB()
        scores = cross_val_score(model, X_scaled, bal_y, cv=cross_val)
        model.fit(X_scaled, bal_y)
        pred_gaussian = model.predict(X_scaled)
        gaussian_result = classification_report(bal_y, pred_gaussian)
        gaussian_confmatr = confusion_matrix(bal_y, pred_gaussian)
        return scores, gaussian_result, gaussian_confmatr
    
    def DecisionTree(X_scaled, bal_y):
        from sklearn.tree import DecisionTreeClassifier
        from sklearn.model_selection import cross_val_score
        from sklearn.model_selection import StratifiedKFold
        import warnings
        warnings.simplefilter(action='ignore', category=FutureWarning)
        
        model = DecisionTreeClassifier()
        cross_val = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        scores = cross_val_score(model, X_scaled, bal_y, cv=cross_val)
        model.fit(X_scaled, bal_y)
        pred_decisiontree = model.predict(X_scaled)
        decisiontreee_result = classification_report(bal_y, pred_decisiontree)
        decisiontree_confmatr = confusion_matrix(bal_y, pred_decisiontree)
        return scores, decisiontreee_result, decisiontree_confmatr
    
    a1, a2, a3 =  logReg(X_scaled, bal_y)  
    print('################## LOGISTIC REGRESSION #######################')
    print('Cross-validation scores: ', a1)
    print('Mean score: ', a1.mean())
    print('Standard deviation: ', a1.std())
    print(a2)
    print('Logistic confusion Matrix:')
    print(a3)
    
    b1, b2, b3 = svc(X_scaled, bal_y)
    print('################## SUPPORT VECTOR MACHINE #######################')
    print('Cross-validation scores: ', b1)
    print('Mean score: ', b1.mean())
    print('Standard deviation: ', b1.std())
    print('Support Vector Classification Report:')
    print(b2)
    print('Support Vector Confusion Matrix:')
    print(b3)

    c1, c2, c3 = KNN(X_scaled, bal_y)
    print('################## K NEAREST NEIGHBOR #######################')
    print('Cross-validation scores: ', c1)
    print('Mean score: ', c1.mean())
    print('Standard deviation: ', c1.std())
    print('Knearest Classification Report:')
    print(c2)
    print('Knearest Confusion Matrix:')
    print(c3)

    d1, d2, d3 = Gaussian(X_scaled, bal_y)
    print('################## GAUSSIAN #######################')
    print('Cross-validation scores: ', d1)
    print('Mean score: ', d1.mean())
    print('Standard deviation: ', d1.std())
    print('Gaussian Classification Report:')
    print(d2)
    print('Gaussian Confusion Matrix:')
    print(d3)
    
    e1, e2, e3 = DecisionTree(X_scaled, bal_y)
    print('################## DECISION TREE #######################')
    print('Cross-validation scores: ', e1)
    print('Mean score: ', e1.mean())
    print('Standard deviation: ', e1.std())
    print('DescisionTree Classification Report:')
    print(e2)  
    print('DescisionTree Confusion Matrix:')
    print(e3) 
```
